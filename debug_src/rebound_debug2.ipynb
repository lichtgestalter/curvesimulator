```python
import rebound
import math

# defining some constants
g     = 6.67430e-11          # [m**3/kg/s**2] gravitational constant
m_sun = 1.98847e30           # [kg]  solar mass
m_jup = 1.8981246e27         # [kg]  Jupiter mass
deg   = 1 / 180 * math.pi
day   = 24 * 60 * 60         # [s]
T0    = 2458400.0            # [d]  Epoch
TTc00 = 2458401.4086         # [d]  observed TT Number  0 of TOI-4504c
TTc31 = 2460970.0865         # [d]  observed TT Number 31 of TOI-4504c

# setting up the simulation
sim2 = rebound.Simulation()
sim2.G = g

name = "TOI-4504"   # star
mass = 0.885 * m_sun           # [kg]
sim2.add(hash=name, m=mass)

name = "TOI-4504d"  # inner warm Jupiter
mass =      2.1512    * m_jup  #  [kg]
e =         0.2502    * 1      #  [1]
i =         88.76     * deg    #  [rad]
P =         41.7819   * day    #  [s]
Omega =     0.0       * deg    #  [rad]
omega =     276.77    * deg    #  [rad]
ma =        337.16    * deg    #  [rad]
sim2.add(hash=name, m=mass, P=P, inc=i, e=e, Omega=Omega, omega=omega, M=ma)

name = "TOI-4504c"  # outer warm Jupiter
mass =      2.6385    * m_jup  #  [kg]
e =         0.0350    * 1      #  [1]
i =         89.71     * deg    #  [rad]
P =         81.820097 * day    #  [s]
Omega =    -2.016956  * deg    #  [rad]
omega =     303.62    * deg    #  [rad]
ma =        137.89    * deg    #  [rad]
sim2.add(hash=name, m=mass, P=P, inc=i, e=e, Omega=Omega, omega=omega, M=ma)

sim2.move_to_com()

t = (TTc31 - T0) * day  #  [s] Seconds since Epoch
sim2.integrate(t)

print(f"dx at last transit of TOI4504c at {TTc31}: {sim2.particles[2].x - sim2.particles[0].x:.0e} m (Should me much smaller than 1e+08)")

print("The printed value should be much smaller.")
print("Something is wrong in this use of Rebound.")
print("Do all parameters of the add method have the units, which rebound is expecting?")
print("Am I using the correct parameter names when calling add?")
print("Does rebound interprete the parameters as Jacobian in this setup?")
print("Is seconds the correct unit for the parameter of the integrate method?")
```
