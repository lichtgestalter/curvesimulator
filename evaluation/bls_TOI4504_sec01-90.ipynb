```python
df = csv2df(path+"all_p.csv")
# df = csv2df(path+"27/27_SPOC_120_p.csv")
# df = csv2df(path+"90/90_SPOC_120_p.csv")
plot_this(df.time, [df.flux], ["flux"], title="All processed TOI4504 flux")
plot_this(df.time, [df.flux_err], ["flux_err"], title="All processed TOI4504 flux error Zoom Transits 1-13", left=2458350, right=2458600)
plot_this(df.time, [df.flux_err], ["flux_err"], title="All processed TOI4504 flux error Zoom Transits 27-38", left=2459000, right=2459400)
df
```

```python
bls = BoxLeastSquares(t=df.time, y=df.flux, dy=df.flux_err)
```

```python
durations = np.linspace(0.30, 0.35, 200)  # c T14 0.32 - 0.34
periods = np.linspace(81.0, 84.5, 2000)
results = bls.power(periods, durations)
periodogram(results)
print("actual c period varies between 81.8 and 83.6")
```

```python
durations = np.linspace(0.05, 0.15, 200)  # d 0.05 - 0.15
periods = np.linspace(8.0, 80.0, 2000)
results = bls.power(periods, durations)
periodogram(results)
print("Slight changes of the upper and lower limit of the period intervall decide whether the peak is at 41 days.")
```

```python
durations = np.linspace(0.05, 0.1, 200)  # b T14 0.094 - 0.0942
periods = np.linspace(2.0, 3.0, 2000)
results = bls.power(periods, durations)
periodogram(results)
print("b-Transits are clearly there, if the period search space is small enough. Here 2.0 - 3.0")
print("These transits vanish for example for 0.2 - 3.0 or 2.0 - 30.0")
```
