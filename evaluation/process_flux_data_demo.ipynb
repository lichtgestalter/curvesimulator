# Demo
This notebook's only purpose is to demonstrate the use of the functions defined in process_flux_data.py

```python
# we want to see plots right here in the notebook
%matplotlib inline
from process_flux_data import *  # process_flux_data.py must be in the same directory as this notebook
# import lightkurve as lk   # uncomment if you want to use lightkurve methods in this notebook
path = '../research/star_systems/TOI-4504/lightkurve/'  # path to example lightcurve data. Change this if required.
```

```python
df = csv2df(path+"1/1_GSFC-ELEANOR-LITE_1800.csv")  # path and file name of example lightcurve data. Change this if required.
plot_this(df.time, [df.flux], ["flux"], title="Original Data")
df
```

```python
df = extract_from_df(df, 1326, 1327)
plot_this(df.time, [df.flux], ["flux"], title="Extract")
df
```

```python
df = remove_from_df(df, 1326.1, 1326.9)
plot_this(df.time, [df.flux], ["flux"], title="Removed data in the middle")
df
```

```python
df = scale_flux(df, 5)
plot_this(df.time, [df.flux], ["flux"], title="Flux scaled by 5")
df
```

```python
df = tesstime2bjd(df)
plot_this(df.time, [df.flux], ["flux"], "Time converted to full BJD days")
df
```

```python
df = bjd2tess_time(df)
df = scale_flux(df, 0.2)
plot_this(df.time, [df.flux], ["flux"], title="Undone scaling and conversion to full BJD days")
df
```

```python
median = median_flux(df)
df = scale_flux(df, 1/median)
plot_this(df.time, [df.flux], ["flux"], title="Normalized to median=1")
df
```

```python
print("Convert data frame into LightCurve object if you want to use methods from the Lightkurve library.")
lc = df2lc(df)
lc.plot(label="Lightcurve")
```

```python
# df = lc2df(lc)
df = pd.DataFrame({'time': lc.time.value, 'flux': lc.flux, 'flux_err': lc.flux_err})
plot_this(df.time, [df.flux], ["flux"], title="Converted LightCurve to DataFrame again")
df
```
