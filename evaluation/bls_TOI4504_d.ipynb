# BLS on TOI4504: Where are the d-transits?


```python
# we want to see plots right here in the notebook
%matplotlib inline
from process_TESS_data.process_flux_data import *  # process_flux_data.py must be in the same directory as this notebook
# import lightkurve as lk   # uncomment if you want to use lightkurve methods in this notebook
import astropy
from astropy.timeseries import BoxLeastSquares
from matplotlib import rcParams
path = '../research/star_systems/TOI-4504/lightkurve/'  # path to lightcurve data
```

```python
durations = np.linspace(0.01, 0.15, 200)  # d 0.05 - 0.15
periods = np.linspace(10.0, 90.0, 2000)
```

```python
df = csv2df(path+"all_p.csv")
df = remove_c_transits(df, 0.175)
bls01_90 = BoxLeastSquares(t=df.time, y=df.flux, dy=df.flux_err)
results = bls01_90.power(periods, durations)
print("Sector 1 - 90:")
periodogram(results)
```

```python
index = np.argmax(results.power)
period = results.period[index]
t0 = results.transit_time[index]
duration = results.duration[index]

# Plot the light curve and best-fit model
fig, ax = plt.subplots(1, 1, figsize=(6, 3))
ax.plot(df.time, df.flux, ".k", ms=1)
x = np.linspace(df.time.min(), df.time.max(), 3 * len(df.time))
model_flux = bls01_90.model(x, period, duration, t0)
ax.plot(x, model_flux, lw=2)
ax.set_xlim(df.time.min(), df.time.max())
ax.set_xlabel("time [days]")
ax.set_ylabel("flux [e-/s]")
```
