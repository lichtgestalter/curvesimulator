```python
# we want to see plots right here in the notebook
%matplotlib inline
from process_TESS_data.process_flux_data import *  # process_flux_data.py must be in the same directory as this notebook
# import lightkurve as lk   # uncomment if you want to use lightkurve methods in this notebook
import astropy
from astropy.timeseries import BoxLeastSquares
from matplotlib import rcParams
path = '../../research/star_systems/TOI-4504/lightkurve/'  # path to lightcurve data
```

```python
df = csv2df(path+"27-38_p.csv")
# df = csv2df(path+"27/27_SPOC_120_p.csv")
# df = csv2df(path+"90/90_SPOC_120_p.csv")
plot_this(df.time, [df.flux], ["flux"], title="All processed TOI4504 flux")
df
```

```python
bls = BoxLeastSquares(t=df.time, y=df.flux, dy=df.flux_err)
```

```python
durations = np.linspace(0.30, 0.35, 200)  # c T14 0.32 - 0.34
periods = np.linspace(81.0, 85.5, 2000)
results = bls.power(periods, durations)
periodogram(results)
print("actual c period varies between 81.8 and 83.6")
```

```python
durations = np.linspace(0.05, 0.15, 200)  # d 0.05 - 0.15
periods = np.linspace(24.0, 70.0, 2000)
results = bls.power(periods, durations)
periodogram(results)
print("Slight changes of the upper and lower limit of the period intervall decide whether the peak is at 41 days.")
```
